```python
import numpy as np
import pandas as pd
import sklearn
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
print(movies.shape)
print(credits.shape)
movies.head(2)
```

```python
credits.head()
```

```python
credits['cast'].head(1).values
```

```python
# merge datas

movies = movies.merge(credits, on='title')
```

```python
movies.head(3)
```

```python
# generes
# keywords
# title
# overview
# production_companies    
# cast                  
# crew            
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.info()
```

```python
movies.fillna('', inplace= True)
```

```python
# Check for duplicate Rows

movies.duplicated().sum()
```

```python
movies['genres'][0]

# ['Action', 'Adventure', 'Fantasy', 'SciFi'] is to be obtained
```

```python
# Convert string to list
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies['genres'][0]
```

```python
movies['cast'][0]
```

```python

# first 3 names
def convert2(obj):
    L=[]
    j =0
    for i in ast.literal_eval(obj):
        if j < 3:
            L.append(i['name'])
            j+=1
        else:
            break
    return L  
        
    
```

```python
movies['cast'] = movies['cast'].apply(convert2)
movies['cast'][0]
```

```python
movies.head()
```

```python
movies['crew'][1]
```

```python
# fetch director_name from the list in crew column

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
        
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head(4)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(4)
```

```python
def remove_space(col):
    movies[col] = movies[col].apply(lambda x:[i.replace(' ','') for i in x])
    print(movies[col].head())
```

```python
remove_space('crew')
```

```python
remove_space('keywords')
```

```python
remove_space('cast')
```

```python
remove_space('genres')
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head(3)
```

```python
# Join the words in the list with space in between to form a string

new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
# Text Vectorization is used to convert text into vector.
# Each movie becomes a vector, vectors are plotted on Graph
# we are using Bag of words technique
# Sumup all the tags and extract 5000 most recurring words from the text.
# make a table consisting of all 5000 words and make a column to mention the frequency of the words
# Consider each of the movie tag and find out the count of every word in the tag.

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words= 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
# 5000D vector is formed in which each movie is represented as the vector.
```

```python
len(cv.get_feature_names())
```

```python
vectors[5]
```

```python
cv.get_feature_names()
```

```python
# Stemming;

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Stemming is applied on each of the words of the text
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
ps.stem(new_df['tags'][2])
```

```python

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Angle between each of the vectors is found and used for relating the movies.
# Cosine of angle is used.
# As cosine increases, similarity decreases

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]

# Cos 0 = 1.
```

```python
# Add indicies beside of every distance, and sort the distances.
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
